### 1.1 Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площад

```python
df['sqft_lot'] = pd.to_numeric(df['sqft_lot'], errors='coerce') # Преобразовать данные в столбцах в числовые
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
 
print('Стоимость кв.м в первых 10-ти домах:\n', df['price_per_sq_lot'].head(10))
```
